```python
from learning_spaces.kst import iita, simu, iita_exclude_transitive
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def show_graph(implications):
    G = nx.DiGraph()
    G.add_edges_from(implications)
    nx.draw_networkx(G)
    plt.show()
```
