```python
from learning_spaces.kst import iita, simu, iita_exclude_transitive
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
G.add_edges_from(response['implications'])
nx.draw_networkx(G)
plt.show()
```
