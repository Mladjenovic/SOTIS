```python
from learning_spaces.kst import iita
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
data_frame = pd.DataFrame({
    'p0': [1, 1, 1, 1, 1, 1, 1, 0],
    'p1': [1, 1, 1, 1, 0, 0, 0, 0],
    'p2': [1, 1, 0, 0, 0, 0, 0, 0],
    'p3': [0, 0, 1, 1, 0, 0, 0, 0]
    })
response = iita(data_frame, v=1)
print(response)
```

```python
G = nx.DiGraph()
G.add_edges_from(response['implications'])
nx.draw_networkx(G)
plt.show()
```
